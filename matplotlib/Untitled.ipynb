```python
import matplotlib.pyplot as plt
x=[1,2.5,3,4.6,5,6,4.5]
y=[7,6.5,8.3,5,4.2,4.7,5.7]
x1=[3,2.9,7,4.5,1,9,7.5]
y1=[8,6.9,3.3,2,3.2,6.7,7.8]
plt.scatter(x,y,label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high saving',color='k')
plt.xlabel("saving")
plt.ylabel("income")
plt.title("scatter plot")
plt.legend()
```

```python
import matplotlib.pyplot as plt
days=[1,2,3]
python=[7,8,9]
math=[2,3,4]
bio=[7,8,7]
chem=[8,5,7]
slices=[6,4,7,7]
activities=['python','math','bio','chem']
cols=['c','m','r','b']
plt.pie(slices,
        labels=activities,
        startangle=90,
        shadow=True,
        explode=(0,0.1,0,0),
        autopct='%1.1f%%')

plt.title('pie plot')
plt.show()
```
